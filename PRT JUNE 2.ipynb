```
def is_armstrong_number(number):
    """
    Checks if a number is an Armstrong number.
    An Armstrong number is a number that is equal to the sum of its own digits
    each raised to the power of the number of digits.
    """
    num_str = str(number)
    num_digits = len(num_str)
    sum_of_powers = 0
    for digit_char in num_str:
        sum_of_powers += int(digit_char) ** num_digits
    return sum_of_powers == number

def generate_armstrong_numbers(limit):
    """
    Generates all Armstrong numbers up to a given limit.
    """
    armstrong_numbers = []
    for i in range(1, limit + 1):
        if is_armstrong_number(i):
            armstrong_numbers.append(i)
    return armstrong_numbers
```

```
# Example usage:
limit = 1000
armstrong_nums = generate_armstrong_numbers(limit)
print(f"Armstrong numbers up to {limit}: {armstrong_nums}")
```

```
def sum_digits_arithmetic(number):
    """
    Calculates the sum of the digits of a given number using only arithmetic operations.
    """
    if number < 0:
        number = abs(number) # Handle negative numbers by taking absolute value
    sum_val = 0
    while number > 0:
        sum_val += number % 10  # Add the last digit
        number //= 10           # Remove the last digit
    return sum_val

def is_harshad_number(number):
    """
    Determines if a number is a Harshad (Niven) number.
    A Harshad number is an integer that is divisible by the sum of its digits.
    """
    if number <= 0:
        return False # Harshad numbers are typically positive integers

    digit_sum = sum_digits_arithmetic(number)
    return number % digit_sum == 0
```

```
# Example usage:
num1 = 156
sum1 = sum_digits_arithmetic(num1)
print(f"The sum of digits of {num1} is: {sum1}")
print(f"{num1} is a Harshad number: {is_harshad_number(num1)}")

num2 = 17
sum2 = sum_digits_arithmetic(num2)
print(f"\nThe sum of digits of {num2} is: {sum2}")
print(f"{num2} is a Harshad number: {is_harshad_number(num2)}")

num3 = 21
sum3 = sum_digits_arithmetic(num3)
print(f"\nThe sum of digits of {num3} is: {sum3}")
print(f"{num3} is a Harshad number: {is_harshad_number(num3)}")
```

```
class Loan:
    def __init__(self, principal, annual_interest_rate, time_years):
        """
        Initializes the Loan object with principal amount, annual interest rate, and time in years.

        Args:
            principal (float): The initial amount of money borrowed or invested.
            annual_interest_rate (float): The annual interest rate (e.g., 0.05 for 5%).
            time_years (int): The duration of the loan in years.
        """
        if not all(isinstance(arg, (int, float)) and arg >= 0 for arg in [principal, annual_interest_rate, time_years]):
            raise ValueError("Principal, annual interest rate, and time must be non-negative numbers.")

        self.principal = principal
        self.annual_interest_rate = annual_interest_rate
        self.time_years = time_years

    def calculate_simple_interest(self):
        """
        Calculates the simple interest for the loan.
        Formula: Simple Interest = Principal * Rate * Time

        Returns:
            float: The calculated simple interest.
        """
        simple_interest = self.principal * self.annual_interest_rate * self.time_years
        return simple_interest

    def calculate_total_amount(self):
        """
        Calculates the total amount to be paid back (Principal + Simple Interest).

        Returns:
            float: The total amount after adding simple interest.
        """
        total_amount = self.principal + self.calculate_simple_interest()
        return total_amount
```

```
# Example usage:

# Create a Loan object
loan1 = Loan(principal=10000, annual_interest_rate=0.05, time_years=3)

# Calculate simple interest
simple_interest = loan1.calculate_simple_interest()
print(f"Principal: ${loan1.principal:.2f}")
print(f"Annual Interest Rate: {loan1.annual_interest_rate*100:.2f}%")
print(f"Time (Years): {loan1.time_years}")
print(f"Simple Interest: ${simple_interest:.2f}")

# Calculate total amount
total_amount = loan1.calculate_total_amount()
print(f"Total Amount to be Paid: ${total_amount:.2f}")

# Another example
loan2 = Loan(principal=500, annual_interest_rate=0.10, time_years=1)
print(f"\nPrincipal: ${loan2.principal:.2f}")
print(f"Annual Interest Rate: {loan2.annual_interest_rate*100:.2f}%")
print(f"Time (Years): {loan2.time_years}")
print(f"Simple Interest: ${loan2.calculate_simple_interest():.2f}")
print(f"Total Amount to be Paid: ${loan2.calculate_total_amount():.2f}")
```
